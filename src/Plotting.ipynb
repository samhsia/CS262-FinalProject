```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('accuracies_baseline.npy', 'rb') as f:
    baseline = np.load(f)
    
with open('accuracies_1agent_1noise.npy', 'rb') as f:
    a1n1 = np.load(f)
    
with open('accuracies_1agent_1noise_ad.npy', 'rb') as f:
    a1n1ad = np.load(f)
    
with open('accuracies_1agent_5noise.npy', 'rb') as f:
    a1n5 = np.load(f)

with open('accuracies_1agent_5noise_ad.npy', 'rb') as f:
    a1n5ad = np.load(f)
    
with open('accuracies_1agent_10noise.npy', 'rb') as f:
    a1n10 = np.load(f)
    
with open('accuracies_1agent_10noise_ad.npy', 'rb') as f:
    a1n10ad = np.load(f)

with open('accuracies_3agent_10noise.npy', 'rb') as f:
    a3n10 = np.load(f)
    
with open('accuracies_3agent_10noise_ad.npy', 'rb') as f:
    a3n10ad = np.load(f)
    
with open('accuracies_5agent_10noise.npy', 'rb') as f:
    a5n10 = np.load(f)
    
with open('accuracies_5agent_10noise_ad.npy', 'rb') as f:
    a5n10ad = np.load(f)
```

```python
plt.figure(figsize=(10,5), dpi=442)

plt.plot(np.arange(1, 101), np.mean(baseline, axis=1), label = 'Baseline (no noise)')
plt.plot(np.arange(1, 101), np.mean(a1n1, axis=1), label = 'Noise ~U(0,1)')
plt.plot(np.arange(1, 101), np.mean(a1n5, axis=1), label = 'Noise ~U(0,5)')
plt.plot(np.arange(1, 101), np.mean(a1n10, axis=1), label = 'Noise ~U(0,10)')

plt.xlabel('Number of Training Rounds')
plt.ylabel('Mean Model Accuracy')

plt.xlim([1,100])
plt.ylim([0,100])

plt.legend()

plt.savefig('figures/baseline.png')
plt.show()
```

```python
plt.figure(figsize=(10,5), dpi=442)

plt.plot(np.arange(1, 101), np.mean(baseline, axis=1), label = 'Baseline (no noise)')
plt.plot(np.arange(1, 101), np.mean(a1n10, axis=1), label = '1 Agent')
plt.plot(np.arange(1, 101), np.mean(a3n10, axis=1), label = '3 Agents')
plt.plot(np.arange(1, 101), np.mean(a5n10, axis=1), label = '5 Agents')

plt.xlabel('Number of Training Rounds')
plt.ylabel('Mean Model Accuracy')

plt.xlim([1,100])
plt.ylim([0,100])

plt.legend()

plt.savefig('figures/num_agents.png')
plt.show()
```

```python
plt.figure(figsize=(10,5), dpi=442)

plt.bar(0, np.mean(baseline[-1]),
        width = 0.8, edgecolor='black', color='tab:blue', label = 'Baseline (no noise)')

plt.bar(1, np.mean(a1n1[-1]),
        width = 0.8, edgecolor='black', color='tab:red', label = 'Noise')
plt.bar(2, np.mean(a1n1ad[-1]),
        width = 0.8, edgecolor='black', color='tab:green', label = 'Noise + Recovery')

plt.bar(3, np.mean(a1n5[-1]),
        width = 0.8, edgecolor='black', color='tab:red')
plt.bar(4, np.mean(a1n5ad[-1]),
        width = 0.8, edgecolor='black', color='tab:green')

plt.bar(5, np.mean(a1n10[-1]),
        width = 0.8, edgecolor='black', color='tab:red')
plt.bar(6, np.mean(a1n10ad[-1]),
        width = 0.8, edgecolor='black', color='tab:green')

plt.bar(7, np.mean(a3n10[-1]),
        width = 0.8, edgecolor='black', color='tab:red')
plt.bar(8, np.mean(a3n10ad[-1]),
        width = 0.8, edgecolor='black', color='tab:green')

plt.bar(9, np.mean(a5n10[-1]),
        width = 0.8, edgecolor='black', color='tab:red')
plt.bar(10, np.mean(a5n10ad[-1]),
        width = 0.8, edgecolor='black', color='tab:green')

plt.ylabel('Mean Final Model Accuracy')
plt.xticks([0, 1.5, 3.5, 5.5, 7.5, 9.5], 
           ['Baseline', '1 Agent ~U(0,1)', '1 Agent ~U(0,5)', 
           '1 Agent ~U(0,10)', '3 Agents ~U(0,10)', '5 Agents ~U(0,10)'], rotation=15)

plt.xlim([-1,11])
plt.ylim([0,100])

plt.legend()

plt.savefig('figures/anomaly_detection.png')
plt.show()
```
